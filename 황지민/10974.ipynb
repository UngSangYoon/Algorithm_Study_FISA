```python
# 2503


# 서로 다른 숫자 세 개 묻기
# 위치 및 숫자 일치: 스트라이크
# 숫자 일치 : 볼

import itertools

N = int(input())
q = []
result_count = 0

for _ in range(N):
    num, s, b = input().split()
    q.append((num, int(s), int(b)))

for candidate in itertools.permutations('123456789', 3):
    candidate = ''.join(candidate)
    valid = True
    
    for num, s, b in q:
        strikes  = sum(candidate[j] == num[j] for j in range(3))
        balls  = sum(candidate[j] in num for j in range(3)) - strikes
    
        if strikes  != s or balls != b:
            valid = False
            break

    if valid:
        result_count += 1

print(result_count)
```

```python
# 질문의 개수를 입력받음
N = int(input().strip())
queries = []

# 각 질문(세 자리 수, 스트라이크, 볼)을 입력받아 리스트에 저장
for _ in range(N):
    parts = input().split()
    num, s, b = parts[0], int(parts[1]), int(parts[2])
    queries.append((num, s, b))

result_count = 0

# 1부터 9까지의 숫자 중 서로 다른 3자리 수 후보 생성 (라이브러리 없이 3중 for문 사용)
for i in range(1, 10):
    for j in range(1, 10):
        if j == i:
            continue  # 중복 방지
        for k in range(1, 10):
            if k == i or k == j:
                continue  # 중복 방지
            candidate = str(i) + str(j) + str(k)  # 후보 생성 (예: "123")
            valid = True  # 이 후보가 모든 질문을 만족하는지 확인하는 플래그

            # 각 질문에 대해 후보가 맞는지 검사
            for num, s, b in queries:
                # 스트라이크 계산: 같은 자리에서 같은 숫자인 경우
                strikes = 0
                if candidate[0] == num[0]:
                    strikes += 1
                if candidate[1] == num[1]:
                    strikes += 1
                if candidate[2] == num[2]:
                    strikes += 1

                # 볼 계산: 후보의 숫자가 num에 포함되어 있으나 자리수가 다르면 볼
                total_common = 0
                if candidate[0] in num:
                    total_common += 1
                if candidate[1] in num:
                    total_common += 1
                if candidate[2] in num:
                    total_common += 1

                balls = total_common - strikes

                # 질문에서 주어진 스트라이크(s)와 볼(b)와 다르면 후보 배제
                if strikes != s or balls != b:
                    valid = False
                    break  # 하나의 질문이라도 만족하지 않으면 더 검사할 필요 없음

            # 모든 질문을 만족하면 후보 개수 증가
            if valid:
                result_count += 1

# 가능한 후보의 총 개수를 출력
print(result_count)
```

```python
import requests
import os
from datetime import datetime
from dotenv import load_dotenv
from datetime import datetime
load_dotenv()
today = datetime.today().strftime("%Y%m%d")
API_KEY = os.getenv("AUTHKEY")
URL = f"https://www.koreaexim.go.kr/site/program/financial/exchangeJSON?{API_KEY}&searchdate={today}&data=AP01"


# README 파일 경로
README_PATH = "README.md"

def get_weather():
    """koreaexim API를 호출하여 서울의 날씨 데이터를 가져옴"""
    response = requests.get(URL)
    if response.status_code == 200:
        data = response.json()
        ttb = data["main"]["ttb"]
        cur_nm = data["cur_nm"][0]["description"]
        tts = data["main"]["tts"]
        return f"국가/통화명	: {cur_nm}, 전신환(송금)받으실때: {ttb}°C, 전신환(송금)보내실때: {tts}%"
    else:
        return "날씨 정보를 가져오는 데 실패했습니다."

def update_readme():
    """README.md 파일을 업데이트"""
    weather_info = get_weather()
    now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    readme_content = f"""

# Koreaexim API Status

이 리포지토리는 Koreaexim API를 사용하여 환율 정보를 자동으로 업데이트합니다.

## 현재 환율
> {weather_info}

⏳ 업데이트 시간: {now} (UTC)

---
자동 업데이트 봇에 의해 관리됩니다.
"""

    with open(README_PATH, "w", encoding="utf-8") as file:
        file.write(readme_content)

if __name__ == "__main__":
    update_readme()

```
