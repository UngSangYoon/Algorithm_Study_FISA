```python
# 2503


# 서로 다른 숫자 세 개 묻기
# 위치 및 숫자 일치: 스트라이크
# 숫자 일치 : 볼

import itertools

N = int(input())
q = []
result_count = 0

for _ in range(N):
    num, s, b = input().split()
    q.append((num, int(s), int(b)))

for candidate in itertools.permutations('123456789', 3):
    candidate = ''.join(candidate)
    valid = True
    
    for num, s, b in q:
        strikes  = sum(candidate[j] == num[j] for j in range(3))
        balls  = sum(candidate[j] in num for j in range(3)) - strikes
    
        if strikes  != s or balls != b:
            valid = False
            break

    if valid:
        result_count += 1

print(result_count)
```

```python
# 질문의 개수를 입력받음
N = int(input().strip())
queries = []

# 각 질문(세 자리 수, 스트라이크, 볼)을 입력받아 리스트에 저장
for _ in range(N):
    parts = input().split()
    num, s, b = parts[0], int(parts[1]), int(parts[2])
    queries.append((num, s, b))

result_count = 0

# 1부터 9까지의 숫자 중 서로 다른 3자리 수 후보 생성 (라이브러리 없이 3중 for문 사용)
for i in range(1, 10):
    for j in range(1, 10):
        if j == i:
            continue  # 중복 방지
        for k in range(1, 10):
            if k == i or k == j:
                continue  # 중복 방지
            candidate = str(i) + str(j) + str(k)  # 후보 생성 (예: "123")
            valid = True  # 이 후보가 모든 질문을 만족하는지 확인하는 플래그

            # 각 질문에 대해 후보가 맞는지 검사
            for num, s, b in queries:
                # 스트라이크 계산: 같은 자리에서 같은 숫자인 경우
                strikes = 0
                if candidate[0] == num[0]:
                    strikes += 1
                if candidate[1] == num[1]:
                    strikes += 1
                if candidate[2] == num[2]:
                    strikes += 1

                # 볼 계산: 후보의 숫자가 num에 포함되어 있으나 자리수가 다르면 볼
                total_common = 0
                if candidate[0] in num:
                    total_common += 1
                if candidate[1] in num:
                    total_common += 1
                if candidate[2] in num:
                    total_common += 1

                balls = total_common - strikes

                # 질문에서 주어진 스트라이크(s)와 볼(b)와 다르면 후보 배제
                if strikes != s or balls != b:
                    valid = False
                    break  # 하나의 질문이라도 만족하지 않으면 더 검사할 필요 없음

            # 모든 질문을 만족하면 후보 개수 증가
            if valid:
                result_count += 1

# 가능한 후보의 총 개수를 출력
print(result_count)
```

```python
# 1149

'''
예제 입력 1 
3
26 40 83
49 60 57
13 89 99
예제 출력 1 
96
예제 입력 2 
3
1 100 100
100 1 100
100 100 1
예제 출력 2 
3
'''

# 집 N개
# 집은 순서대로 있음
# 빨강, 초록, 파란 중 하나의 색
# 모든 집을 칠하는 비용의 최솟값
# 1 != 2
# N != N-1
# i( 2 <= i <= N-1) != i-1, i+1

N = int(input())

cost = [map(int, input().split()) for _ in range(N)]

f = [0] * (N+1) # DP Table

for i in range(N):
    f[i+1] = cost[i][0]

for i in range(2, N -1):
    f[i] = min(f[i-1], f[i-3])
```
